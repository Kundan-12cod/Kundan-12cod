```python
#9.  Write a Python program to apply RandomizedSearchCV for tuning hyperparameters (C, penalty, solver) in
#Logistic Regression. Print the best parameters and accuracy
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
# Load dataset
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Define parameter grid
param_dist = {
    'C': np.logspace(-4, 4, 20),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}


```

```python
# Apply RandomizedSearchCV
random_search = RandomizedSearchCV(LogisticRegression(), param_dist, n_iter=10, cv=5, random_state=42)
random_search.fit(X_train, y_train)
```

```python
# Print best parameters and accuracy
print("Best Parameters:", random_search.best_params_)
y_pred = random_search.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#15.  Write a Python program to apply feature scaling (Standardization) before training a Logistic Regression model. Evaluate its accuracy and compare results with and without scaling
# Load dataset
from sklearn.preprocessing import StandardScaler

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Apply Standardization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Logistic Regression with and without scaling
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy without scaling:", accuracy_score(y_test, y_pred))
```

```python
model.fit(X_train_scaled, y_train)
y_pred_scaled = model.predict(X_test_scaled)
print("Accuracy with scaling:", accuracy_score(y_test, y_pred_scaled))
```

```python
#20. Write a Python program to train Logistic Regression and visualize the Precision-Recall Curve for binary classificatio:
from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import label_binarize
# Train Logistic Regression
model = LogisticRegression(multi_class='ovr')
model.fit(X_train, y_train)
```

```python
# Binarize the output for precision_recall_curve
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
y_pred_proba = model.predict_proba(X_test)
```

```python
# Plot Precision-Recall Curve for each class
for i in range(3): # Iterate through each class
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_pred_proba[:, i])
    plt.plot(recall, precision, marker='.', label=f'Class {i}')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend() # Add a legend to differentiate classes
plt.show()
```
