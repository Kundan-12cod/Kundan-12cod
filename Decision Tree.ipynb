```python
# 16. Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy  
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,classification_report
from sklearn.model_selection import train_test_split



```

```python
X,y = load_iris(return_X_y= True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30 ,random_state= 1)
```

```python
# Train Decision Tree Classifier
model = DecisionTreeClassifier()
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred  = model.predict(X_test)
print("accuracy :" , accuracy_score(y_pred,y_test))
```

```python
#19 Write a Python program to train a Decision Tree Regressor on a housing dataset and evaluate using Mean Squared Error (MSE)


from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

```

```python
# Load dataset
X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train Decision Tree Regressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
# Evaluate using MSE
y_pred = model.predict(X_test)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
#27. Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy
from sklearn.model_selection import cross_val_score
# Find optimal alpha for pruning
path = model.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
```

```python
# Train models with different alpha values
accuracies = []
for alpha in ccp_alphas:
    model = DecisionTreeClassifier(ccp_alpha=alpha)
    scores = cross_val_score(model, X_train, y_train, cv=5)
    accuracies.append(scores.mean())
```

```python
# Plot accuracy vs alpha
plt.plot(ccp_alphas, accuracies)
plt.xlabel("Alpha")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Alpha for Cost Complexity Pruning")
plt.show()
```
